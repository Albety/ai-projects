```python
import torch
```

### Logit

```python
x = torch.tensor([0.0, 1.0, 1.0])

w = torch.tensor([0.1, 0.5, 1.0])

b = torch.tensor(-0.7)

weighted_sum = torch.dot(w, x)

z= weighted_sum + b


print(f"Взвешенная сумма до смещения: {weighted_sum.item()}")
print(f"Итоговый выход нейрона (z): {z.item()}")
```

### Activation functions

```python
import torch.nn.functional as F

z_tensor = torch.tensor([-2.5, 0.5, 3.0])

relu_output = F.relu(z_tensor)

sigmoid_output = torch.sigmoid(z_tensor)





print(f"Сырой сигнал (z): {z_tensor}")
print(f"После ReLU:       {relu_output}")
print(f"После Sigmoid:    {sigmoid_output}")
```
